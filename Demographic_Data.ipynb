Q5 What percenatage of people without advanced education make more than 50k?

```python
lower_education=(df.education!='Bachelors') & (df.education !='Master') & (df.education !='Doctorate')
lower_education.index.name='Lower_education'
pd.DataFrame(lower_education).head()
```

```python
lo_ed_hi_salary = df.loc[lower_education & (df['salary'] == '>50K')].value_counts().sum()
lo_ed_total = df.loc[lower_education].value_counts().sum()
print("low Ed Hi Salary",lo_ed_hi_salary)
print("low Ed Total",lo_ed_total)
```

```python
#ANS5
Lower_education_hi_salary=(lo_ed_hi_salary *100 /lo_ed_total)
Lower_education_hi_salary=round(Lower_education_hi_salary,2)
print('Percentage of Lower Education make more than 50k:',Lower_education_hi_salary)
```
